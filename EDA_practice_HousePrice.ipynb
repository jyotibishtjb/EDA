#sns with "tips" inbuilt dataset in seaborn

```python
df= sns.load_dataset("tips")
df.head()
```

```python
sns.jointplot(x='tip',y='total_bill', data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.distplot(df['tip'])
```

```python
#categorical plots
#1. count plot
sns.countplot(x='tip', data=df)
plt.figure(figsize=(20,10))
sns.displot(x='tip', data=df)
plt.tight_layout()
plt.show()
```
